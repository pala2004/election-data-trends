## Filter for 2014 data - Uttar Pradesh

```python
data_2014 = df[df['year'] == 2014]
up_data_2014 = data_2014[data_2014['st_name'] =='Uttar Pradesh']
up_data_2014_max_voted_party = up_data_2014.groupby('partyabbre')['totvotpoll'].sum()
up_data_2014_max_voted_party = pd.DataFrame(up_data_2014_max_voted_party)
up_data_2014_max_voted_party = up_data_2014_max_voted_party.sort_values(by= 'totvotpoll',ascending=False)
up_data_2014_top5_max_voted_party = up_data_2014_max_voted_party.head(5)
up_data_2014_top5_max_voted_party = up_data_2014_top5_max_voted_party.reset_index()
up_data_2014_top5_max_voted_party
```

## Top 5 Parties - Uttar Pradesh 2014

```python
plt.figure(figsize = (12,5))
plt.bar(up_data_2014_top5_max_voted_party['partyabbre'],up_data_2014_top5_max_voted_party['totvotpoll'], 
        color=[ 'orange', 'green', 'blue', 'cyan', 'yellow'])
```

## Distribution of Total Votes Polled

```python
# Analyzing distribution of votes by party
plt.figure(figsize=(12, 5))
sns.histplot(df['totvotpoll'], bins=30, kde=True)
plt.title('Distribution of Total Votes Polled')
plt.xlabel('Total Votes Polled')
plt.ylabel('Frequency')
plt.show()
```

## Total Votes Over the Years

```python
# Analyzing votes over the years
votes_per_year = df.groupby('year')['totvotpoll'].sum().reset_index()
plt.figure(figsize=(12, 5))
sns.lineplot(data=votes_per_year, x='year', y='totvotpoll')
plt.title('Total Votes Polled Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Votes Polled')
plt.show()
```

## Votes by State

```python
# Analyzing votes by state
votes_by_state = df.groupby('st_name')['totvotpoll'].sum().reset_index()
votes_by_state = votes_by_state.sort_values(by='totvotpoll', ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(data=votes_by_state, x='totvotpoll', y='st_name', palette='viridis')
plt.title('Total Votes Polled by State')
plt.xlabel('Total Votes Polled')
plt.ylabel('State')
plt.show()
```

## Gender Distribution of Candidates

```python
# Analyzing gender distribution of candidates
gender_distribution = df['cand_sex'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_distribution.index, y=gender_distribution.values, palette='pastel')
plt.title('Gender Distribution of Candidates')
plt.xlabel('Gender')
plt.ylabel('Number of Candidates')
plt.show()
```

## Votes vs Electors

```python
# Analyzing the relationship between total votes polled and number of electors
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='electors', y='totvotpoll', alpha=0.5)
plt.title('Total Votes Polled vs Number of Electors')
plt.xlabel('Number of Electors')
plt.ylabel('Total Votes Polled')
plt.show()
```

```python
# Voter Turnout Percentage
df['voter_turnout'] = (df['totvotpoll'] / df['electors']) * 100

# Party-wise Vote Share (within each year)
df['total_votes_year'] = df.groupby('year')['totvotpoll'].transform('sum')
df['party_vote_share'] = (df['totvotpoll'] / df['total_votes_year']) * 100
```

## Turnout Over the Years

```python
# Voter Turnout Over the Years
turnout_per_year = df.groupby('year')['voter_turnout'].mean().reset_index()
plt.figure(figsize=(12, 5))
sns.lineplot(data=turnout_per_year, x='year', y='voter_turnout')
plt.title('Average Voter Turnout Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Voter Turnout Percentage')
plt.show()
```

## Top 10 Parties by Vote Share - 2014

```python
# Party-wise Vote Share Analysis (for 2014)
vote_share_2014 = df[df['year'] == 2014].groupby('partyabbre')['party_vote_share'].mean().reset_index()
vote_share_2014 = vote_share_2014.sort_values(by='party_vote_share', ascending=False).head(10)

plt.figure(figsize=(12, 5))
sns.barplot(data=vote_share_2014, x='partyabbre', y='party_vote_share', palette='Set3')
plt.title('Top 10 Parties by Vote Share in 2014')
plt.xlabel('Party Abbreviation')
plt.ylabel('Average Vote Share Percentage')
plt.show()
```

## Performance of Major Parties Over Years

```python
# Comparison of Major Parties' Performance (BJP, INC, SP) Over the Years
major_parties = ['BJP', 'INC', 'SP']
major_parties_data = df[df['partyabbre'].isin(major_parties)]
performance_over_years = major_parties_data.groupby(['year', 'partyabbre'])['totvotpoll'].sum().reset_index()

plt.figure(figsize=(12, 5))
sns.lineplot(data=performance_over_years, x='year', y='totvotpoll', hue='partyabbre', marker='o')
plt.title('Performance of Major Parties Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Votes Polled')
plt.legend(title='Party')
plt.show()
```

## Votes by Candidate Gender (Boxplot)

```python
# Relationship between Candidate's Gender and Votes Received
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='cand_sex', y='totvotpoll', palette='pastel')
plt.title('Distribution of Votes Polled by Gender of Candidate')
plt.xlabel('Gender')
plt.ylabel('Total Votes Polled')
plt.show()
```

## Voter Turnout vs Votes

```python
# Analyzing the relationship between voter turnout and total votes polled
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='voter_turnout', y='totvotpoll', alpha=0.5)
plt.title('Voter Turnout vs Total Votes Polled')
plt.xlabel('Voter Turnout Percentage')
plt.ylabel('Total Votes Polled')
plt.show()
```

## Electors vs Voter Turnout

```python
# Analyzing the relationship between electors and voter turnout
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='electors', y='voter_turnout', alpha=0.5)
plt.title('Number of Electors vs Voter Turnout Percentage')
plt.xlabel('Number of Electors')
plt.ylabel('Voter Turnout Percentage')
plt.show()
```
